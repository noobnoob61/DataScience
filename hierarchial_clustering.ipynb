# HIERARCHICAL CLUSTERING (agglomerative)

Dendrograms :- memory of the algorithm

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#not any dependent variable
dataset=pd.read_csv("Mall_Customers.csv")
X=dataset.iloc[:,[3,4]].values
```

Using Dendrogram to find Optimal number of Clusters

```python
import scipy.cluster.hierarchy as sch 
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('customers')
plt.ylabel('distance')
plt.show()
```

Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering

hc=AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
y_hc
```

Visualising the Clusters

```python
plt.scatter(X[y_hc==0,0], X[y_hc==0,1],s=100,c='red',label='cluster 1')   #spending score column
#s=size
#c=color
plt.scatter(X[y_hc==1,0], X[y_hc==1,1],s=100,c='orange',label='cluster 2') 
plt.scatter(X[y_hc==2,0], X[y_hc==2,1],s=100,c='pink',label='cluster 3') 
plt.scatter(X[y_hc==3,0], X[y_hc==3,1],s=100,c='blue',label='cluster 4') 
plt.scatter(X[y_hc==4,0], X[y_hc==4,1],s=100,c='yellow',label='cluster 5')

plt.title('Clusters of customers')
plt.xlabel('Annual income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
