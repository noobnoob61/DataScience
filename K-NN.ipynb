# K-NN (K NEAREST NEIGHBOUR)

Steps:-
1. Choose the no. K of neighbours
2. Take the K nearest neighbours of new data point according to euclidean distance
3. among these K neighbours, count the number of data points in each category 
4. assign the new data point to the category where you counted the most neighbours 

Your model is ready!

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
X=dataset.iloc[:,:-1].values 
y=dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
y_test=np.array(y_test)
```

Training KNN model on Training Set

1.n_neighbors=5

When a new data point is classified, the algorithm looks at the 5 closest data points (neighbors) in the training set.

2. weights='uniform':

Suppose you have k=5 neighbors.
If 3 neighbors belong to class A and 2 belong to class B, the new sample will be classified as class A because A has the majority of votes.

3.metric='minkowski'

The metric parameter specifies the distance metric used to determine the "closeness" between data points.

minkowski is a generalized distance metric that includes both Euclidean distance and Manhattan distance as special cases, depending on the value of p.

4. p=2

If p=1, the Minkowski distance becomes the Manhattan distance (also known as L1 norm).

If p=2, the Minkowski distance becomes the Euclidean distance (also known as L2 norm).

```python

from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
```

Predicting New result

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

Predicting the Test results

```python
y_pred=classifier.predict(X_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_pred,y_test)

cm
```

Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
