# CLASSIFICATION

### LOGISTIC REGRESSION

1. for one use cases / independent variable (ex- age):-
     ln p/(1-p) = h0 + h2X1

2. for multiple use cases / independent variables (ex- age, income, education, family or single):-
    ln p/(1-p) = h0 + h2X1 +h2X2 + h3X3 + h4X4

Logistic Regression plot graph based on probabilities between EX:- NO(0) OR YES(1) (used in most of the cases of binary outcome)
  

BEST CURVE<= MAX LIKELIHOOD

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
X=dataset.iloc[:,:-1].values 
y=dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test=np.array(y_test)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

Training Logistic Regression on Training Set

```python
from sklearn.linear_model import LogisticRegression

classifier=LogisticRegression()
classifier.fit(X_train,y_train)
```

Predicting New Result

```python
classifier.predict(sc.transform([[30,87000]]))
```

```python
y_pred=classifier.predict(X_test)


print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
classifier.score(X_test,y_test)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

The previous method using classifier.score(y_pred.reshape(1,-1),y_test.reshape(1,-1)) didn't work because the score method in scikit-learn is designed to evaluate the model's performance on input features (X) and corresponding labels (y)

Why We Use accuracy_score:
accuracy_score(y_test, y_pred): This function directly compares the predicted labels (y_pred) with the actual labels (y_test) to compute accuracy. It doesn't require the original feature data (X_test) because it works with the predictions already made.

MAKING THE CONFUSION MATRIX
