# Polynomial Regression


y = b0 + b1x1 +b b2(x1)^2 +... + bn(x1)^n

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Position_Salaries.csv")
X=dataset.iloc[:,1:-1].values 
y=dataset.iloc[:,-1]
```

```python
X  #dont need to include first column as it is redunant and we already have encoding of this in second column as ranks from 1 to 10
```

```python
y
```

Training Linear Regression model

```python
from  sklearn.linear_model import LinearRegression

r1= LinearRegression()
r1.fit(X,y)
```

Training Polynomial Regression model

```python
from sklearn.preprocessing import PolynomialFeatures

r2=PolynomialFeatures(degree=4)
#fit_tranform , transforms the sigle feature i.e second column into matrix feature which is x1, x1^2 as degree 2
X_poly=r2.fit_transform(X)

#After transforming features into matrix of x and x^2, we will make the linear equation finally 
r3=LinearRegression()
r3.fit(X_poly,y)
```

Visualising Linear Regression Results

```python
plt.scatter(X,y,color='red') 
plt.plot(X,r1.predict(X))  #predicted salaries
plt.title('Truth or Bluff')
plt.xlabel("position")
plt.ylabel("salary")
```

Visualising Polynomial Regression results

```python
plt.scatter(X,y,color="purple")
plt.plot(X,r3.predict(X_poly))
plt.title('Truth or Bluff')
plt.xlabel("position")
plt.ylabel("salary")

#increase the power for smoother curve
```

Predicting Salary of 6.5 rank with Linear Regression

```python
r1.predict([[6.5]])  #array = [[]]  first bracket= 1D second= 2D
```

Predicting Salary of 6.5 rank with Polynomial Regression

```python
r3.predict(r2.fit_transform([[6.5]]))  #saying truth
```
